```python
import pandas as pd
import subprocess
import json

def analyze_sentiment_with_ollama(text):
    # Call the Ollama model for sentiment analysis
    result = subprocess.run(['ollama', 'run', 'llama', text], capture_output=True, text=True)
    # It is expected that this output is a JSON format, adjust if incorrect
    return json.loads(result.stdout)

def analyze_sentiments(input_csv, output_csv):
    # Load sentences from the CSV file
    df = pd.read_csv(input_csv)

    # Check if there are at least three columns
    if df.shape[1] < 3:
        raise ValueError("CSV must contain at least three columns.")

    # Initialize lists for sentiment results
    sentiments_column_1 = []
    sentiments_column_2 = []
    sentiments_column_3 = []

    # Analyze sentiment for each of the last three columns
    for _, row in df.iterrows():
        for col_index in range(-3, 0):  # Last three columns
            sentence = row.iloc[col_index]
            if not isinstance(sentence, str):  # Ensure the sentence is a string
                sentiment = "Invalid"
                sentiment_score = 0.0  # Default score for invalid input
            else:
                result = analyze_sentiment_with_ollama(sentence)
                # Extract sentiment information
                sentiment = result.get('sentiment', 'Unknown')  # Adjust according to your output structure
                sentiment_score = result.get('score', 0.0)      # Adjust according to your output structure

            if col_index == -1:
                sentiments_column_1.append((sentiment, sentiment_score))
            elif col_index == -2:
                sentiments_column_2.append((sentiment, sentiment_score))
            elif col_index == -3:
                sentiments_column_3.append((sentiment, sentiment_score))

    # Create new columns in the DataFrame
    df['Sentiment_Column_1'] = [sent[0] for sent in sentiments_column_1]
    df['Sentiment_Score_Column_1'] = [sent[1] for sent in sentiments_column_1]

    df['Sentiment_Column_2'] = [sent[0] for sent in sentiments_column_2]
    df['Sentiment_Score_Column_2'] = [sent[1] for sent in sentiments_column_2]

    df['Sentiment_Column_3'] = [sent[0] for sent in sentiments_column_3]
    df['Sentiment_Score_Column_3'] = [sent[1] for sent in sentiments_column_3]

    # Save the updated DataFrame to a new CSV file
    df.to_csv(output_csv, index=False)

# Example usage
input_csv = 'Course.csv'  # Update this to your actual input CSV file
output_csv = 'output_sentiments.csv'
analyze_sentiments(input_csv, output_csv)
```

```python
import pandas as pd
from googletrans import Translator
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Ensure the VADER lexicon is downloaded
nltk.download('vader_lexicon')

def sentiment_analysis(text):
    """Performs sentiment analysis on the given text."""
    translator = Translator()

    # Translate Arabic text to English
    translated_text = translator.translate(text, src='ar', dest='en').text

    analyzer = SentimentIntensityAnalyzer()
    sentiment = analyzer.polarity_scores(translated_text)

    if sentiment['compound'] >= 0.05:
        return "Positive"
    elif sentiment['compound'] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

def process_csv(input_file, output_file):
    # Read sentences from the input CSV file
    df = pd.read_csv(input_file)

    # Check if there are at least 3 columns
    num_columns = df.shape[1]
    if num_columns < 3:
        raise ValueError("The input CSV must contain at least 3 columns.")

    # Extract the last three columns
    last_three_columns = df.iloc[:, -3:]

    # Initialize sentiment columns
    sentiment_columns = []

    for i in range(last_three_columns.shape[1]):
        # Combine sentences from the current column
        combined_sentences = last_three_columns.iloc[:, i].astype(str)

        # Perform sentiment analysis and store results in a new column
        sentiment_column_name = f'sentiment_{i+1}'
        df[sentiment_column_name] = combined_sentences.apply(sentiment_analysis)
        sentiment_columns.append(sentiment_column_name)

    # Write results to the output CSV file
    df.to_csv(output_file, index=False, encoding='utf-8-sig')

# Example usage
input_file = 'Course.csv'
output_file = 'output1.csv'
process_csv(input_file, output_file)
```
